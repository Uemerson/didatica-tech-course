```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# generating a mass of data
x, y = make_regression(n_samples=200, n_features=1, noise=30)

# showing on the chart
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
# creating the model
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# showing the results
import numpy as np
plt.scatter(x,y)
xreg = np.arange(-4, 4, 1)
angular_coefficient = model.coef_
linear_coefficient = model.intercept_

# y = mx + b
# regression graph
plt.plot(xreg, angular_coefficient*xreg+linear_coefficient, color='red')
plt.show()
```
